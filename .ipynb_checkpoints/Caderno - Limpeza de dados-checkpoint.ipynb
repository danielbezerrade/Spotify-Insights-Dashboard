```python
# Importando as bibliotecas que serão utilizadas

import pandas as pd
```

```python
# Fazendo as configurações dentro da biblioteca

pd.set_option("display.max_columns",None) # Isso mostra todas as colunas, quando chamarmos a tabela (DataFrame)
```

```python
#Definindo a varíavel "dados"
dados = pd.read_csv("Spotify most streamed.csv",encoding="utf-8")
```

```python
#Chamando o nosso DataFrame

dados.head()
```

```python
#Realizando a análise exploratória 

dados.shape
```

```python
dados.dtypes
```

```python
dados.info()
```

```python
#Verificando a quantidade de linhas com valores nulos
dados.isnull().sum()
```

#### **Dados nulos**

Identificamos que em nosso data Frame , possuímos alguns valores nulos, ou seja, valores que não existem ou não vieram na base de dados.
Iremos realizar as tratativas, mais para frente.

```python
#Verificando se há duplicidade

dados.duplicated().sum()
```

```python
#REALIZANDO OS PRIMEIROS AJUSTES NA TABELA , SEPARANDO POR COLUNA "ARTISTA" E "MÚSICA"

dados[["Artist","Song"]] = dados["Artist and Title"].str.split(" - ",n=1,expand=True)
```

```python
dados.head()
```

```python
#Excluindo a primeira coluna

dados.drop(columns="Artist and Title",axis=1,inplace=True)

```

```python
dados.head()
```

```python
#Definindo a ordem das colunas

dados = dados[["Artist","Song","Streams","Daily"]]
```

```python
#Verificando o nosso DataFrame

dados.head()
```

#### **Ajustando as informações do nosso DataFrame**

Como o nosso DataFrame é grande, possuindo muitas linhas, iremos deixar alguns ajustes, padronizando as informações.

```python
#Ajustes na coluna Artist

dados["Artist"] = dados["Artist"].str.strip().str.title() #Tirando os espaços e padronizando as informações, padronizando sempre o ínicio da palavra com maiúsculo
dados.head()
```

```python
dados["Song"] = dados["Song"].str.strip().str.title()
dados.head()
```

```python
#Modificando o tipo de dados da coluna "Streams" - a mesma está em formato de string

print(dados["Streams"].dtype) # o resultado traz "objetc"
dados["Streams"] = pd.to_numeric(dados["Streams"].str.replace(",",""),errors="coerce").astype(float)
```

```python
dados.head()
```

```python
#alterando o tipo da coluna "Streams"

dados["Streams"] = dados["Streams"].astype("Int64")
```

```python
dados.head()
```

#### **Verificando A coluna "Daily"**

Como havíamos verificado, a coluna Daily possui alguns valores nulos. Iremos seguir com o tratamento desses valores.

```python
#verificando os valores nulos

filtro = dados[dados["Daily"].isna()]
print(filtro[["Artist","Song","Streams","Daily"]])
```

```python
#Outro jeito de verificar

filtro2 = dados.loc[dados["Daily"].isna(),["Artist","Song","Streams","Daily"]]
filtro2
```

#### **Realizando a limpeza da coluna "Daily"**

Como não temos a quantidade de reproduções diárias para esses artistas, e faltam algumas informações, vamos remover essas linhas da nossa base de dados.

```python
#Retirando essas linhas que não possuem informação "Daily"

dados = dados.dropna(subset=["Daily"])
```

```python
#Verificando agora

dados.isnull().sum()
```

```python
#Formatando a coluna "Daily".Mudando o tipo 

print(dados["Daily"].dtype) # O resultado é "object"

dados["Daily"] = pd.to_numeric(dados["Daily"].str.replace(",","").astype("Int64"))
```

#### **Como sabemos que o nosso DataFrame não é uma cópia, por hora, iremos ignorar esse aviso.**

```python
print(dados["Daily"].dtype)
```
